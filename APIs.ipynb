####**What is an Application Programming Interafce (API)**?

An interface between two applicatoions or serveers to allow communication and exchange of data and functionality using a simple set of commands, and without the need to code the functionality from scratch

**Example**: Github API, uber, weather forecasting apps, E-commerce payment gates


######**Advantages of using APIs**:
* Data sharing (e.g. airlines sharing flight times with a travel app)
* App integrations (e.g. using Gmail to sign into an app)
* Embedded content (e.g. latest 5 tweets on a blog page)



######**Why are APIs important for data scientists?**
* Updated and reliable data
* Model deployment
* Scalibility

---

####**What are the types of APIs?**

APIs are categorized based on access and architecture.

**Based on access:**
* **Open API**: can be accessed by any developer
* **Partner API**: shared among businesses only
* **Internal API**: accessed by members of a team at a company
* **Composite API**: combine multiple APIs

**Based on architecture:**
* **Simple Object Access Protocol(SOAP)**: it's a communuication protocol, and rely on HTTP and XML for exchanging data

* **Representational state transfer(REST)**: utilizes JSON and CSV to encode information; uses HTTPS requests for communications. These requests are:


>**GET**: retrieve data

> **POST**: add new data

> **DELETE**: remove data

> **PUT**: update data

RESTFUL APIs need an endpoint! `Endpoints` for APIs are a specific and unique URL to communicate with a specific resource in the server. These are defined by the server we're trying to communicate with, and can be found in the API documentation for that server. 

*Example: Get users who have liked a Tweet - `requests.get(/2/tweets/:id/liking_users, header=api)`*



---


**Most web APIs follow a similar structure of implementation:**
1. Generate/apply for your API key
2. Authentication in CLI or your program
3. Depending on the module used, get the necessary endpoints (e.g.`Requests` for RESTful APIs require an endpoint)
4. Review the API documentation
5. Review the module documentation
